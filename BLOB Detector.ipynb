```python
import cv2
import matplotlib.pylab as plt 

img = cv2.imread(r"C:\Users\zbook 17 g3\Desktop\TrOCR handwriting analysis-\Sunflower-fields-and-mazes-in-Ontario-toronto-mississauga.png")

params = cv2.SimpleBlobDetector_Params()
params.minThreshold = 40
params.maxThreshold = 255

params.filterByArea = True
params.minArea = 10000
params.maxArea = 100000

params.filterByColor = False
params.blobColor = 0

params.filterByConvexity = True
params.minConvexity = .3
params.maxConvexity = 1

detector = cv2.SimpleBlobDetector_create(params)
Key = detector.detect(img)
cir = cv2.drawKeypoints(img, Key, None, (255, 0, 0), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
#cv2.imshow("circles", cir)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
plt.imshow(cv2.cvtColor(cir,cv2.COLOR_BGR2RGB))
plt.title("BLOB Detector")
plt.axis("off")
```
